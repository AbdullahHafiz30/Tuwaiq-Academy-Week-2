```python
import sklearn
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

np.random.seed(18)
```

```python
X = 5 * np.random.rand(50, 1)

y = 4 + 2 * X + np.random.rand(50, 1)
```

```python
plt.plot(X, y ,"b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([0,6,3,18])
plt.show()
```

```python
X_new = 5 + 3 * np.random.rand(50, 1)
```

```python
X_new
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
print('Linear Regression model intercept is', lin_reg.intercept_)

print('Linear Regression model co-effecient is', lin_reg.coef_)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

y_pred = lin_reg.predict(X_train)

mse = mean_squared_error(y_train, y_pred)
print('Mean squared error = ', mse)

mae = mean_absolute_error(y_train, y_pred)
print('Mean absolute error = ', mae)

rmse = np.sqrt(mean_squared_error(y_train, y_pred))
print('Root mean squared error = ', rmse)
```

```python
plt.plot(X, y ,"b.")
plt.plot(X_train, y_pred, "r")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([0,5,0,15])
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

y_pred = lin_reg.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print('Mean squared error = ', mse)

mae = mean_absolute_error(y_test, y_pred)
print('Mean absolute error = ', mae)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print('Root mean squared error = ', rmse)
```

```python
plt.plot(X_test, y_test ,"b.")
plt.plot(X_test, y_pred, "r")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([0,5,0,15])
plt.show()
```

```python
y_pred = lin_reg.predict(X_new)

mse = mean_squared_error(X_new, y_pred)
print('Mean squared error = ', mse)

mae = mean_absolute_error(X_new, y_pred)
print('Mean absolute error = ', mae)

rmse = np.sqrt(mean_squared_error(X_new, y_pred))
print('Root mean squared error = ', rmse)
```
